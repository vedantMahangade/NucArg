```python
import pandas as pd
from Bio import SeqIO
```

```python
fasta_file = "/Users/vedantmahangade/Projects/compbio_project/data/negative classes long read.fasta"
```

```python
long_read_negative = []
```

```python
for record in SeqIO.parse(fasta_file, "fasta"):
    fasta_id = record.id.split(' ')[0]
    dna_seq = str(record.seq)
    db = "NCBI"
    sequence = ""
    type = "non_resistant"
    long_read_negative.append({
        'id': fasta_id,
        'db': db,
        'type': type,
        'sequence': sequence,
        'dna_seq': dna_seq
    })
```

```python
negative_classes = pd.DataFrame(long_read_negative)
```

```python
negative_classes
```

```python
negative_classes.to_csv('long_read_negative.csv', index=False)
```
